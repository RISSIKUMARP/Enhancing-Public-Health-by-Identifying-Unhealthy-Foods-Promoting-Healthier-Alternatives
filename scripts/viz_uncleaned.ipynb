Data Visualization for Uncleaned data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\kavir\OneDrive\Desktop\new\datsets\enriched_kroger_usda_products.csv")
```

```python
# this below code is to get the distribution of calories in products
plt.figure(figsize=(8, 5))
plt.hist(df['calories'].dropna(), bins=30, edgecolor='black')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.title('Distribution of Calories in Products')
plt.show()
```

```python
# this one describes a Boxplot of the overall Macronutrients distribution
plt.figure(figsize=(8, 5))
df[['protein', 'fat', 'carbohydrates']].boxplot()
plt.ylabel('Grams per Serving')
plt.title('Macronutrient Distribution')
plt.show()
```

```python
#to check the frequency of healthy and unhealthy products
calorie_threshold = 200 
protein_threshold = 5    
fat_threshold = 10      

df['health_status'] = df.apply(
    lambda row: 'Healthy' if (row['calories'] <= calorie_threshold and 
                              row['protein'] >= protein_threshold and 
                              row['fat'] <= fat_threshold) else 'Unhealthy',
    axis=1
)

health_status_counts = df['health_status'].value_counts()

plt.figure(figsize=(8, 5))
health_status_counts.plot(kind='bar', color=['green', 'red'], edgecolor='black', alpha=0.7)
plt.xlabel('Health Status')
plt.ylabel('Frequency')
plt.title('Frequency of Healthy vs Unhealthy Products')
plt.xticks(rotation=0)
plt.show()
```

```python
# This demonstrates a Histogram of Protein Content
plt.figure(figsize=(8, 5))
plt.hist(df['protein'].dropna(), bins=30, edgecolor='black')
plt.xlabel('Protein (g)')
plt.ylabel('Frequency')
plt.title('Distribution of Protein in Products')
plt.show()

```

```python
# Using a boxplot to visualize fat content in healthy and unhealthy products
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='health_status', y='fat', palette={'Healthy': 'green', 'Unhealthy': 'red'})
plt.xlabel('Health Status')
plt.ylabel('Fat (g)')
plt.title('Boxplot of Fat Content in Healthy and Unhealthy Products')
plt.show()
```
