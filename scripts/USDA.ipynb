```python
import requests

API_KEY = "oC14rRpGrTUUKwkbwQeBk0eISyuolUo8BIXW6V5C"  #got this API key signing in to USDA

# Base URL for USDA FoodData Central API
BASE_URL = "https://api.nal.usda.gov/fdc/v1/foods/search"

def get_food_data(food_name, max_results=5):
    
    #Fetching food nutrition data from USDA FoodData Central API.

    
    params = {
        "query": food_name,
        "pageSize": max_results,
        "api_key": API_KEY
    }
    
    response = requests.get(BASE_URL, params=params)
    
    if response.status_code == 200:
        data = response.json()
        return data["foods"]
    else:
        print(f"Error {response.status_code}: {response.text}")
        return None

# Example code to see if it is fetching the right details by choosing the term apple
food_name = "apple"
foods = get_food_data(food_name)

if foods:
    for i, food in enumerate(foods):
        print(f"\nResult {i+1}: {food['description']}")
        for nutrient in food["foodNutrients"][:5]:  # Show first 5 nutrients
            print(f" - {nutrient['nutrientName']}: {nutrient['value']} {nutrient['unitName']}")
else:
    print("No results found.")
```
